```python
# Notes:
    # General Shape
    # variable_name[start:end:step]
    # start: the index to start slicing from (inclusive — will appear in output)
    # end: the index to stop slicing at (exclusive — will NOT appear in output)
    # step: is the value of steps will be walked
    # if step = 1: you will walk one by one, without skipping any elements
    # if step = 2: you will take every second element (skip one between) one and so one ...
    # the sign of step determines direction:
    # - step > 0 → walk left to right (forward)
    # - step < 0 → walk right to left (backward)

    # if start or end are negative:
    # - python first converts them to their positive index equivalent (from the right side)
    # - always resolve negative indices to be sure what you're doing
    # core Logic Rules:
    # if step > 0:
    # - if start >= end → empty (no output)
    # - if start <  end → output
    # if step < 0:
    # - if start > end → output
    # - if start <= end → empty (no output)
    # these rules apply AFTER converting any negative indices
    # this logic works for all sequence types: strings, lists, tuples, etc.
    # be careful: if any index is negative, you can't be sure just by looking.
    # → always convert it to its equivalent positive index before solving
    # → understand the logic, don’t just memorize rules blindly
    # if no index for start py will start from 0 by default
    # if no index for end py will stop at the index by default
    # if no step for walking py will walk by one step by default

```
